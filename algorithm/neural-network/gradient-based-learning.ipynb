# Gradient-based Learning

```python
import matplotlib.pyplot as plt
import numpy as np

def f(x): return 1/10 * x**2
def df_dx(x): return 1/5 * x

x = 3 # 초기값 x를 3으로 설정
ITERATIONS = 20 # 총 20번의 update

fig, (ax1, ax2) = plt.subplots(figsize=(10, 7), nrows=2, ncols=1)
_x = np.linspace(start=-5, stop=5, num=100)
_y = f(_x)
ax1.plot(_x, _y)

x_track, y_track = [x], [f(x)]
for iter in range(ITERATIONS):
    dy_dx = df_dx(x)
    x = x - dy_dx
    x_track.append(x)
    y_track.append(f(x))

ax1.scatter(x=x_track, y=y_track, c=range(ITERATIONS+1), cmap="rainbow")
ax2.plot([n for n in range(len(x_track))], x_track)
ax2.scatter(x=[n for n in range(len(x_track))], y=x_track)

df_dx2 = lambda x: 1/5 * (x-2)
for iter in range(ITERATIONS):
    dy_dx = df_dx2(x)
    x = x - dy_dx
    
```

# Learning Rate
$x:=x-a\frac{\Delta y}{\Delta x}$

```python
# f(x1, x2) = x1^2 + x2^2

f = lambda x1, x2: x1**2 + x2**2
dfx = lambda x: 2*x

ITERATIONS = 10

x1 = 3
x1_arr = [x1]

x2 = 3
x2_arr = [x2]

for i in range(ITERATIONS):
    x1 -= 0.3*dfx(x1)
    print(f"x1: {x1}")
print("-------------------------")
for i in range(ITERATIONS):
    x2 -= 0.3*dfx(x2)
    print(f"x2: {x2}")
```

# Bivariate Function Contour Plot

```python
import numpy as np
import matplotlib.pyplot as plt

f = lambda x1, x2: 1/4 * x1**2 + 1/10 * x2**2

dfx1 = lambda x: 1/2 * x
dfx2 = lambda x: 1/5 * x

f_x1 = np.linspace(-5, 5, 100)
f_x2 = np.linspace(-5, 5, 100)

function_X1, function_X2 = np.meshgrid(f_x1, f_x2)
function_Y = np.log(f(function_X1, function_X2))

ITERATIONS = 20

x1 = 3
x1_arr = [x1]
x1_plot = [x1]

x2 = 3
x2_arr = [x2]
x2_plot = [x2]

while True:
    if x1 < 0.001:
        break
    x1 -= 0.7*dfx1(x1)
    x1_plot.append(x1)
    x2 -= 0.7*dfx2(x2)
    x2_plot.append(x2)

x1 = 3
x2 = 3

for i in range(ITERATIONS):
    x1 -= 0.7*dfx1(x1)
    x1_arr.append(x1)
    x2 -= 0.7*dfx2(x2)
    x2_arr.append(x2)
    
fig, ax = plt.subplots(figsize=(5, 5))
ax.contour(function_X1, function_X2, function_Y,
            levels=100, cmap="Reds_r")
ax.plot(x1_plot, x2_plot)
ax.scatter(x1_arr, x2_arr, zorder=2, s=10)
ax.set_xlabel("x", fontsize=15)
ax.set_ylabel("y", fontsize=15)
ax.tick_params(labelsize=15)
fig.tight_layout()
plt.show()
```
