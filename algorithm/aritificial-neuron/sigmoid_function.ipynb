```python
import numpy as np
import matplotlib.pyplot as plt

class Sigmoid:
    def __call__(self, z) -> np.ndarray:
        return 1 / (1 + np.exp(-z))
    
sigmoid = Sigmoid()
print(sigmoid(1)) # 0.73

fig, ax = plt.subplots(figsize=(7, 5))

x = np.linspace(start=-5, stop=5, num=1000)
y = sigmoid(z=x)

ax.plot(x, y)
min_x, max_x = ax.get_xlim()
min_y, max_y = ax.get_ylim()

plt.axhline(0, min_x, max_x, color="lightgrey", linestyle="--")
plt.axvline(0, min_y, max_y, color="lightgrey", linestyle="--")
plt.show()
```
