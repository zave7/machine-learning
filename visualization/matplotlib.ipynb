```python
import numpy as np
import matplotlib.pyplot as plt

t = np.linspace(-4 * PI, 4 * PI, 1000).reshape(1, -1)

sin = np.sin(t)
cos = np.cos(t)
tan = np.tan(t)
feature_data = np.vstack((sin, cos, tan))

title_list = [r"$sin(t)$", r"$cos(t)$", r"$tan(t)$"]
x_ticks = np.arange(-4 * PI, 4*PI+PI, PI)
x_ticklabels = [str(i) + r"$\pi$" for i in range(-4, 5)]

fig, axes = plt.subplots(nrows=3, ncols=1,
                         figsize=(7,10),
                         sharex=True)

for (ax_idx, ax) in enumerate(axes.flat):
    ax.plot(t.flatten(), feature_data[ax_idx])
    ax.set_title(title_list[ax_idx], fontsize=30)
    ax.tick_params(labelsize=20)
    ax.grid()
    if ax_idx == 2:
        ax.set_ylim(bottom=-3, top=3)
fig.subplots_adjust(left=0.1, right=0.95, bottom=0.05, top=0.95)

axes[-1].set_xticks(x_ticks)
axes[-1].set_xticklabels(x_ticklabels)


```

### Violin 1

```python
from sklearn.datasets import load_iris
import matplotlib.pyplot as plt

iris = load_iris()

fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(10, 10))

range_target = range(len(iris.target_names))
feature_names = iris.feature_names
range_feature = range(len(feature_names))
targets = iris.target
feature_data = list(zip(*iris.data))

feature_target_group = [
    [
        [] for _ in range_target
    ] for _ in range_feature
]

for (idx_feature, feat) in enumerate(feature_data):
    for (idx_target, target) in enumerate(targets):
        feature_target_group[idx_feature][target].append(feat[idx_target])

for (idx_feature, axe) in enumerate(axes.flat):
    axe.violinplot(feature_target_group[idx_feature], positions=range_target)
    axe.set_title(feature_names[idx_feature])
    axe.set_xticks(range_target)
    axe.set_xticklabels(iris.target_names)
```

```python
from sklearn.datasets import load_iris
import matplotlib.pyplot as plt
import numpy as np

iris = load_iris()

fig, axes = plt.subplots(nrows=4, ncols=4, figsize=(15, 15))

range_target = range(len(iris.target_names))
feature_names: list[str] = iris.feature_names
range_feature = range(len(feature_names))
targets: list[str] = iris.target
feature_data = list(zip(*iris.data))

feature_target_group: list[list[list[float]]] = [
    [
        [] for _ in range_target
    ] for _ in range_feature
]

for (idx_feature, feat) in enumerate(feature_data):
    for (idx_target, target) in enumerate(targets):
        feature_target_group[idx_feature][target].append(feat[idx_target])

(rows, cols)  = axes.shape

for col in range(cols):
    for row in range(rows):
        axe = axes[row, col]
        x = np.array(feature_target_group[col]).flat
        y = np.array(feature_target_group[row]).flat
        if row == col:
            axe.hist(np.array(x).flat, rwidth=0.9)
        else:
            for i in range_target:
                axe.scatter(x=np.array(x).flat, 
                            y=np.array(y).flat, 
                            c=targets,
                            s=30, alpha=0.7)
                axe.grid()
        if row == rows - 1:
            axe.set_xlabel(feature_names[col], fontsize=15)
        if col == 0:
            axe.set_ylabel(feature_names[row], fontsize=15)
        
plt.show()
```
